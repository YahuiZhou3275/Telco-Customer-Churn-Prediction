# 7. data preprocessing for Model

```python
# (1) scale data
# (2) split the data into X and y
```

```python
# (1)scale the data
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
data_new_scaled = data_new.copy()
data_new_scaled[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']] = scale.fit_transform(data_new_scaled[['SeniorCitizen', 'tenure', 'MonthlyCharges', 'TotalCharges']])
data_new_scaled
```

```python
# (2) split the data into X and y
```

```python
X = data_new_scaled.drop('Churn', axis=1)
y = data_new_scaled['Churn']

from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# 8. basic model building

```python
# import library 
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

from sklearn.metrics import accuracy_score
```

```python
# use Gaussian Naive Bayes: 
gnb = GaussianNB()
gnb.fit(X_train,y_train)
predictions = gnb.predict(X_test)
score_gnb = accuracy_score(y_test, predictions)
print("The accuracy score of Gaussian Naive Bayes is: %2.2f%%" % (score_gnb*100))
```

```python
# use Logistic Regression:
lr = LogisticRegression(max_iter=2000)
lr.fit(X_train,y_train)
predictions = lr.predict(X_test)
score_lr = accuracy_score(y_test, predictions)
print("The accuracy score of Logistic Regression Model is: %2.2f%%" % (score_lr*100))
```

```python
# use Decision Tree
tr = DecisionTreeClassifier()
tr.fit(X_train,y_train)
predictions = tr.predict(X_test)
score_tr = accuracy_score(y_test, predictions)
print("The accuracy score of Decision Tree Model is: %2.2f%%" % (score_tr*100))
```

```python
# use K-Nearest Neighbors
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
predictions = knn.predict(X_test)
score_knn = accuracy_score(y_test, predictions)
print("The accuracy score of K-Nearest Neighbors Model is: %2.2f%%" % (score_knn*100))
```

```python
# use RandomForest
rf = RandomForestClassifier()
rf.fit(X_train,y_train)
predictions = rf.predict(X_test)
score_rf = accuracy_score(y_test, predictions)
print("The accuracy score of Random Forest is: %2.2f%%" % (score_rf*100))
```

```python
# use Support Vector Machine
svm = SVC()
svm.fit(X_train,y_train)
predictions = svm.predict(X_test)
score_svm = accuracy_score(y_test, predictions)
print("The accuracy score of Support Vector Machine Model is: %2.2f%%" % (score_svm*100))
```

```python
# use xgbboost
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(X_train,y_train)
predictions = xgb.predict(X_test)
score_xgb = accuracy_score(y_test, predictions)
print("The accuracy score of XGBoost Model is: %2.2f%%" % (score_xgb*100))
```

```python
# plot to visualize the performance of all models
columns_name = ['GNB','LR','KNN','DT','RF','SVM','XGB']
scores = np.concatenate([score_gnb,score_lr, score_knn,score_tr,score_rf,score_svm,score_xgb],axis=None)*100
score_df = pd.DataFrame(scores, index=columns_name, columns=['accuracy_score'])
score_df
```

```python
plt.plot(score_df)
```

```python
# Comment:
# we can see that Logistic Regression Model 
# does the best job at prediction whether a customer would leave the company or not
```

# 9. model tuning

```python
from sklearn.model_selection import GridSearchCV
```

```python
# simple performance reporting function
def clf_performance(classifier, model_name):
    print(model_name)
    print("Best Score: %2.2f%%" % (classifier.best_score_*100))
    print("Best Parameters: " + str(classifier.best_params_))
```

```python
# logistic regression
lr = LogisticRegression()
param_grid={'max_iter': [2000],
           'penalty': ['l1','l2'],
            'C':np.logspace(-4,4,20),
           'solver': ['liblinear']}
clf_lr = GridSearchCV(lr, param_grid=param_grid,cv=5,verbose=True)
best_clf_lr = clf_lr.fit(X_train, y_train)
clf_performance(best_clf_lr,'Logistic Regression')
score_tuned_lr = best_clf_lr.best_score_
```

```python
# KNN
knn = KNeighborsClassifier()
param_grid={'n_neighbors': [3, 5, 7, 9],
           'weights': ['uniform', 'distance'],
           'algorithm': ['auto','ball_tree','kd_tree'],
           'p': [1,2]}
clf_knn = GridSearchCV(knn, param_grid=param_grid,cv=5,verbose=True)
best_clf_knn = clf_knn.fit(X_train, y_train)
clf_performance(best_clf_knn,'KNN')
score_tuned_knn = best_clf_knn.best_score_
```

```python
# SVM
svm = SVC()
param_grid=[{'kernel': ['rbf'], 'gamma':[.1, .5, 1, 2, 5, 10], 'C':[.1, .5, 1, 2, 5, 10]},
            {'kernel':['linear'], 'C':[.1, 1, 10, 100, 1000]},
            {'kernel':['poly'], 'degree':[2, 3, 4, 5], 'C':[.1, 1, 10, 100, 1000]}]
clf_svm = GridSearchCV(svm, param_grid=param_grid,cv=5,verbose=True)
best_clf_svm = clf_svm.fit(X_train, y_train)
clf_performance(best_clf_svm,'SVM')
score_tuned_svm = best_clf_svm.best_score_
```

```python
# random forest classifier
rftree = RandomForestClassifier()
param_grid = {'n_estimators':[400, 450, 500, 550],
              'criterion':['gini','entropy'],
              'bootstrap':[True],
              'max_depth':[15,20,25],
              'max_features':['auto','sqrt', 10]}
clf_rftree = GridSearchCV(rftree, param_grid=param_grid,cv=5,verbose=True)
best_clf_rftree = clf_rftree.fit(X_train, y_train)
clf_performance(best_clf_rftree,'Random Forest')
score_tuned_rftree = best_clf_rftree.best_score_
```

```python
# xgb
xgb = XGBClassifier()
param_grid = {'n_estimators':[450, 500, 550],
             'colsample_bytree':[0.75, 0.8, 0.85],
             'max_depth':[None],
             'reg_lambda':[2, 5, 10],
             'reg_alpha':[1],
             'subsample':[0.55, 0.6, 0.65],
             'learning_rate':[0.5],
             'gamma':[0.5, 1, 2],
             'min_child_weight':[0.01],
             'sampling_method':['uniform']}
clf_xgb = GridSearchCV(xgb, param_grid=param_grid,cv=5,verbose=True)
best_clf_xgb = clf_xgb.fit(X_train, y_train)
clf_performance(best_clf_xgb,'XGB')
score_tuned_xgb = best_clf_xgb.best_score_
```

```python
# plot to visualize the performance of all models
columns_name = ['GNB','LR','KNN','DT','RF','SVM','XGB']
scores_tuned = np.concatenate([score_tuned_gnb,score_tuned_lr, score_tuned_knn,score_tuned_tr,score_tuned_rftree,score_tuned_svm,score_tuned_xgb],axis=None)*100
score_df = pd.DataFrame(scores_tuned, index=columns_name, columns=['accuracy_score_tuned'])
score_df
```
